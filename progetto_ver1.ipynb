**12) For each year, compute the total amount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st, 2016, planned expiration time January 30th 2018, and amount 5000USD has an amount of 5000USD * 31 / (31+365+30) = 363.85 for 2016, 5000USD * 365 / (31+365+30) = 4284.04 for 2017, and 5000USD * 30 / (31+365+30) = 352.11 for 2018.**

```python
loans_final=loans[['loan_id','disburse_time','planned_expiration_time','duration','loan_amount']].dropna()
```

```python
loans_final['fraction_amount_per_day']=(loans_final['loan_amount']/loans_final['duration']).round(2)
loans_final
```

```python
def create_list(start,end):
    return list(range(start,end+1))
```

```python
loans_final['years']=loans_final.apply(lambda x: create_list(x['disburse_time'].year,x['planned_expiration_time'].year),axis=1)
```

```python
loans_final
```

```python
loans_final=loans_final.explode('years')
loans_final=loans_final.reset_index(drop=True).dropna()
```

```python
def amount_per_year(year, amount_per_day, disburse_time, planned_expiration_time,loan_amount):
    
    if (disburse_time.year == planned_expiration_time.year):
            return loan_amount
        
    elif (year == disburse_time.year):
            end_year=datetime.date(year, 12, 31)
            duration=(end_year-disburse_time.date()).days
            return amount_per_day*duration
        
    elif (year == planned_expiration_time.year):
            start_year=datetime.date(year, 1, 1)
            duration=(planned_expiration_time.date()-start_year).days
            return amount_per_day*duration
        
    else:
        duration=(datetime.date(year, 12, 31) - datetime.date(year, 1, 1)).days   
        return amount_per_day*duration
    
```

```python
loans_final['amount_year']=loans_final.apply(lambda x: amount_per_year(x['years'],x['fraction_amount_per_day'],x['disburse_time'],x['planned_expiration_time'],x['loan_amount']),axis=1)
```

```python
loans_final
```

```python
loans_final.groupby(['years'],as_index=False)['amount_year'].sum().round(1)
```
